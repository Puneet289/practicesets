#DATASET READING

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df=pd.read_csv('/content/drive/My Drive/Datasets/fifa19.csv')
```

```
df.head()
```

```
df.columns
```

#DATA CLEANING

```
df=df.drop(['Unnamed: 0', 'ID', 'Name','Age','Photo', 'Nationality', 'Flag','Club','Club Logo', 'Value', 'Wage','Special','Preferred Foot','Work Rate', 'Body Type', 'Real Face', 'Position','Jersey Number', 'Joined', 'Loaned From', 'Contract Valid Until','Release Clause'],axis=1)
```

```
df.columns
```

```
df.head()
```

```
df.info()
```

```
df=df.dropna()
```

```
df=df.reset_index()
```

```
df.head()
```

```
df.info()
```

```
def addit(col):
  x,y=col.split('+')
  return int(x)+int(y)
```

```
addcols=['LS', 'ST', 'RS', 'LW', 'LF', 'CF',
       'RF', 'RW', 'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB',
       'LDM', 'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']
```

```
for i in addcols:
  df[i]=df[i].apply(addit)
```

```
df.head()
```

```
def height(inp):
  feet,inches=inp.split("'")
  feet=int(feet)
  inches=int(inches)
  feet=feet*12
  result=float(feet+inches)
  result=result*2.54
  return(result)
```

```
def weight(inp):
  inp=inp.replace("lbs","")
  inp=int(inp)
  result=inp/2.205
  return(result)
```

```
df['Height']=df['Height'].apply(height)
```

```
df['Weight']=df['Weight'].apply(weight)
```

#FEATURE SELECTION

```
corr=df.corr()
```

```
plt.figure(figsize=(40,40))
sns.heatmap(corr,cmap='RdYlGn',annot=False,cbar=False,linewidths=1)
```

```
X=df.drop('Overall',axis=1)
```

```
y=df['Overall']
```

```
from sklearn.model_selection import train_test_split
```

```
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=101)
```

#MODEL SELECTION

```
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,BaggingRegressor,VotingRegressor
from sklearn.svm import SVR
```

```
lr=LinearRegression()
dtree=DecisionTreeRegressor()
rfr=RandomForestRegressor(n_estimators=20)
svr=SVR(kernel='poly',degree=2)
adb=AdaBoostRegressor(DecisionTreeRegressor(),n_estimators=10)
bgr=BaggingRegressor(DecisionTreeRegressor(),n_estimators=20)
evr=VotingRegressor([('lr',lr),('dtree',dtree),('svr',svr)])
```

```
lr.fit(X_train,y_train)
```

```
dtree.fit(X_train,y_train)
```

```
rfr.fit(X_train,y_train)
```

```
svr.fit(X_train,y_train)
```

```
adb.fit(X_train,y_train)
```

```
bgr.fit(X_train,y_train)
```

```
evr.fit(X_train,y_train)
```

```
print("Linear Regression:",lr.score(X_train,y_train),"\n","Decision Tree: ",dtree.score(X_train,y_train),"\n","Random Forest: ",rfr.score(X_train,y_train),"\n","Support Vector Machine: ",svr.score(X_train,y_train),"\n","Adaboost Regressor:",adb.score(X_train,y_train),"\n","Bagging Regressor:",bgr.score(X_train,y_train),"\n","Voting Regressor:",evr.score(X_train,y_train))
```

```
#Decision tree and adaboost regressors overfit so we drop them...
```

```
print("Linear Regression:",lr.score(X_test,y_test),"\n","Random Forest: ",rfr.score(X_test,y_test),"\n","Support Vector Machine: ",svr.score(X_test,y_test),"\n","Bagging Regressor:",bgr.score(X_test,y_test),"\n","Voting Regressor:",evr.score(X_test,y_test))
```

Above results state that bagging regressor and random forest provides best accuracy...
